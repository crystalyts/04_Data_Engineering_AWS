```python
def get_weather_loop(cities):

  API_key = '012bf1e02188785784af48c17d0f4eb9'

  tz = pytz.timezone('Europe/Berlin')
  now = datetime.now().astimezone(tz)

  weather_dict = {'city_id': [],
                'country': [],
                'forecast_time': [],
                'outlook': [],
                'weather_description': [],
                'temperature': [],
                'feels_like': [],
                'risk_of_rain': [],
                'rain': [],
                'snow': [],
                'wind_speed': [],
                'humidity': [],
                'information_retrieved_at': []}

  for city in cities:
    url = (f"http://api.openweathermap.org/data/2.5/forecast?q={city}&appid={API_key}&units=metric")
    response = requests.get(url)
    json = response.json()

    for i in json['list']:
        if city == 'Berlin':
            weather_dict['city_id'].append(1)
        elif city == 'Hamburg':
            weather_dict['city_id'].append(2)
        elif city == 'Frankfurt':
            weather_dict['city_id'].append(3)
        elif city == 'Munich':
            weather_dict['city_id'].append(4)
        elif city == 'Stuttgart':
            weather_dict['city_id'].append(5)
        elif city == 'Cologne':
            weather_dict['city_id'].append(6)
        elif city == 'Hannover':
            weather_dict['city_id'].append(7)

        weather_dict['country'].append(json['city']['country'])
        weather_dict['forecast_time'].append(i['dt_txt'])
        weather_dict['outlook'].append(i['weather'][0]['main'])
        weather_dict['weather_description'].append(i['weather'][0]['description'])
        weather_dict['temperature'].append(i['main']['temp'])
        weather_dict['feels_like'].append(i['main']['feels_like'])
        weather_dict['risk_of_rain'].append(i['pop'])
        try:
            weather_dict['rain'].append(i['rain']['3h'])
        except:
            weather_dict['rain'].append('0')
        try:
            weather_dict['snow'].append(i['snow']['3h'])
        except:
           weather_dict['snow'].append('0')
        weather_dict['wind_speed'].append(i['wind']['speed'])
        weather_dict['humidity'].append(i['main']['humidity'])
        weather_dict['information_retrieved_at'].append(now.strftime("%d/%m/%Y %H:%M:%S"))
        cities = ['Berlin', 'Hamburg', 'Frankfurt','Munich','Stuttgart','Cologne','Hannover']
  weather_df = pd.DataFrame(weather_dict)
  return weather_df
```
